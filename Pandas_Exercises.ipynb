```python
from pandas.core.indexes.numeric import Int64Index
def clean_col(name):
  return (name.lower().replace('/','_').replace(' ','_'))

def tweak_df(df):
    return (
            df
            .rename(columns=clean_col)
            .assign(new_household = df["households"].astype(int))

    )  
  
house_price_new = tweak_df(house_price)
house_price_new
```
